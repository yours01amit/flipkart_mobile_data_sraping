# lets make final data


```python
%%time
final = pd.DataFrame()

for i in range(1,1001):
    _url = "https://www.flipkart.com/search?q=" + "mobile"
    webpage = requests.get(_url).text
    soup = BeautifulSoup(webpage,'html.parser')
    mobiles = soup.find_all("div", class_="cPHDOP")

    mobile_ka_name = []
    mobile_ka_price = []
    mobile_ki_ratings = []
    mobile_ke_features = []

        
    for j in mobiles:
        try:
            mobile_ka_name.append(j.find("div",class_="KzDlHZ").text)
        except:
            mobile_ka_name.append(np.nan)

        try:        
            mobile_ka_price.append(j.find("div", class_="Nx9bqj _4b5DiR").text)
        except:
            mobile_ka_price.append(np.nan)

        try:        
            mobile_ki_ratings.append(j.find("div", class_="_5OesEi").text)
        except:
            mobile_ki_ratings.append(np.nan)

        try:        
            mobile_ke_features.append(j.find("div", class_="_5OesEi").text)
        except:     
            mobile_ke_features.append(np.nan)


my_data = {"mobile_name":mobile_ka_name, "price":mobile_ka_price, "Ratings_And_Reviews":mobile_ki_ratings, "features":mobile_ke_features}

df = pd.DataFrame(my_data)

final = pd.concat([final, df], ignore_index=True)
```

```python
final.to_csv("mobile_data.csv")
```
